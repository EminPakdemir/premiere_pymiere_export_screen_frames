```python
app = pymiere.objects.app
SequenceName = app.project.activeSequence.name
print(SequenceName)
os.mkdir(f"{os.getcwd()}\\Export_Frames\\{SequenceName}") # Sekans adında bir klasör açılır.
os.mkdir(f"{os.getcwd()}\\Export_Frames\\{SequenceName}\\Jpg")
os.mkdir(f"{os.getcwd()}\\Export_Frames\\{SequenceName}\\Png")
```

```python

i=0
n=0
f = 22 #f karede 1 atlama yapıcak
workAreaTime = app.project.activeSequence.getWorkAreaOutPointAsTime().seconds
print(workAreaTime)

print(workAreaTime / 0.02)  # toplam frame sayısı

print(workAreaTime / 0.02 / f) # toplam atlama yapılacak sayı

timebase = app.project.activeSequence.timebase # 1 ticks sayısını bulmak için
#print("1 Frame ticks Timebase: ", timebase)

presetPath = f"{os.getcwd()}\\Preset_File\\png_preset.epr"

while i < workAreaTime / 0.02 / f:
    print(i)

    print("Frame: ", n)
    app.project.activeSequence.setInPoint(0.02*n)
    app.project.activeSequence.setOutPoint(0.02*(n+1))

    
    
    #time frame 0.02 bir kare arası zaman...
    app.project.activeSequence.setPlayerPosition(timebase*n)
    #print("Current Frame Time Ticks: ", timebase*n)
    if i <9:
        outputPathPng = f"{os.getcwd()}\\Export_Frames\\{SequenceName}\\Png\\{SequenceName}_0{i+1}.png"
        outputPathJpg = f"{os.getcwd()}\\Export_Frames\\{SequenceName}\\Jpg\\{SequenceName}_0{i+1}.jpg"
        app.project.activeSequence.exportAsMediaDirect(outputPathPng, presetPath, 1)

        #png olarak çıkan resimler jpg olarak convert edilir.
        im = Image.open(outputPathPng)
        rgb_im = im.convert('RGB')
        rgb_im.save(outputPathJpg, quality=80)

       
    else:
        outputPathPng = f"{os.getcwd()}\\Export_Frames\\{SequenceName}\\Png\\{SequenceName}_{i+1}.png"
        outputPathJpg = f"{os.getcwd()}\\Export_Frames\\{SequenceName}\\Jpg\\{SequenceName}_{i+1}.jpg"
        app.project.activeSequence.exportAsMediaDirect(outputPathPng, presetPath, 1)

        #png olarak çıkan resimler jpg olarak convert edilir.
        im = Image.open(outputPathPng)
        rgb_im = im.convert('RGB')
        rgb_im.save(outputPathJpg, quality=80)

    i+=1
    n += f

#app.project.activeSequence.exportAsMediaDirect
#app.project.sequences[index].exportAsMediaDirect(outputPath, presetPath, workAreaType)

#workAreaType
#Must be one of the following:
#0 ENCODE_ENTIRE
#1 ENCODE_IN_TO_OUT
#2 ENCODE_WORK_AREA
```
